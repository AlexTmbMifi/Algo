```python
def binary_search(list, item):
    low = 0
    high = len(list)
    steps = 0
    while low <= high:
        steps += 1
        mid = int((low + high)/ 2)
        guess = list[mid]
        if guess == item:
            return steps
        if guess > item:
            high = mid - 1
        else:
            low = mid + 1
    return None
```

```python
result_steps = []
for i in range(3, 10000):
    a = list(range(i))
    result_steps.append(binary_search(a, 1))
plt.plot(result_steps)
```
